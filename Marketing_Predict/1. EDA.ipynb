## 1. Obtain the data

In this project there is a folder call data where I will use to put all the data, if you want to clone the repository you will not find it, but you can get the dataset from kaggle.

```python
import pandas as pd

# Load the dataset
DATA_FOLDER = "../Marketing_Predict/data/"
FILE_PATH = DATA_FOLDER + "marketing_campaign_dataset.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(FILE_PATH)

# Display the first few rows of the DataFrame
df.head()
```

## 2. Basic Analysis to know the dataset

```python
def basic_analysis(df):
    # Display the shape of the DataFrame
    print("DataFrame Shape:")
    print(df.shape)
    print("\n")
    
    # Check for missing values
    print("Missing Values:")
    print(df.isnull().sum())
    print("\n")

    # Display data types and non-null counts
    print("Data Info:")
    print(df.info())
    print("\n")

# Perform basic analysis on the DataFrame
basic_analysis(df)
```

This dataset don't have nulls and missing values, but it needs to prepare the data.

## 3. Data preparation

```python
# Change the date columns to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Remove dollar sign and commas from 'Acquisition_Cost' column and convert it to float
df['Acquisition_Cost'] = df['Acquisition_Cost'].str.replace('$', '').str.replace(',', '').astype(float)

# Convert 'Duration' column to numeric by extracting the number of days
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(int)

df.info()
```
